```python
#ライブラリのインポート
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_boston
from scipy.stats import zscore
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error
```

```python
#データの取得
from sklearn.datasets import load_boston
boston = load_boston()
```

```python
# DataFrame
#CRIM	人口 1 人当たりの犯罪発生数
#ZN	25,000 平方フィート以上の住居区画の占める割合
#NDUS	小売業以外の商業が占める面積の割合
#CHAS	チャールズ川によるダミー変数 (1: 川の周辺, 0: それ以外)
#NOX	NOx の濃度 窒素化合物
#RM	住居の平均部屋数
#AGE	1940 年より前に建てられた物件の割合
#DIS	5 つのボストン市の雇用施設からの距離 (重み付け済)
#RAD	環状高速道路へのアクセスしやすさ
#TAX	$10,000 ドルあたりの不動産税率の総計
#PTRATIO	町毎の児童と教師の比率
#B	町毎の黒人 (Bk) の比率を次の式で表したもの。 1000(Bk – 0.63)^2
#LSTAT	給与の低い職業に従事する人口の割合 (%)
boston_df = pd.DataFrame(boston.data)
boston_df.columns = boston.feature_names
boston_df
```

```python
#目的変数を追加
boston_df["Price"]=boston.target
boston_df.head()
```

```python
# 説明変数をLSTATのみにする
X =boston_df.LSTAT
X.head()
```

```python
# 目的変数
Y = boston_df.Price
Y.head()
```

```python
#データをテストデータとトレインデータに分ける
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2 , random_state=1)
```

```python
lin_2d.fit(x_train_2,y_train)
lin_3d.fit(x_train_3,y_train)
lin_4d.fit(x_train_4,y_train)
```

```python
score_2d = lin_2d.score(x_test_2,y_test)
score_3d = lin_3d.score(x_test_3,y_test)
score_4d = lin_4d.score(x_test_4,y_test)
print("二次式における'LSTAT'の住宅価格への決定係数は%.2f"%(score_2d))
print("三次式における'LSTAT'の住宅価格への決定係数は%.2f"%(score_3d))
print("四次式における'LSTAT'の住宅価格への決定係数は%.2f"%(score_4d))
```

線形回帰
線形回帰とは、目的変数と説明変数の関係性が線形であるということを仮定した上で、目的変数の値を説明変数の値から予測するような回帰式を求めることである。
具体的には、アイスクリームの売上個数を、その日の天気や温度から予測するなどの問題が挙げられ、在庫管理をしなければならないビジネスの場面でも応用が可能なものである。
回帰と分類の違いには、分類は予測の数値が離散的であるのに対して、回帰では予測の数値が連続的であるという違いがある。

決定係数
推定された回帰式とデータとの当てはまりの良さを示す。値の幅は０〜１であり、１に近づくほど当てはまりの精度が良く、説明変数が目的変数をよく説明できていることを表す。
決定係数を求めるためには、実際のデータとデータ全体の平均値との差である”全変動”、回帰式から得られた予測値とデータ全体の平均値の差である”回帰変動”、実際のデータと回帰式から得られた予測の差である”残差変動”それぞれの二乗和を算出する。全変動は回帰変動と残差変動の和に等しく、決定係数とは、残差変動が全変動に対してどれだけ少ないか（＝回帰変動が全変動に対してどれだけ多いか）を表す指標であるため、回帰変動を前変動で割ることで求められる。この値は相関係数の二乗に等しくなる。
重回帰分析を行う場合、説明変数が多いほど決定係数が高くなる特徴を持つため、目的変数に全く関係ない変数を増やしても決定係数は１に近づいてしまう。そのため、決定係数が高くても、説明変数の数が大きい場合は注意が必要である。これを防ぐために自由度調整済み決定係数を用いることで、説明変数を増やしたことによって決定係数が上がったかの判断ができるようになる。また説明変数が多くなると、説明変数の間で相関が高いものが含まれる（＝多重共線性）可能性があるために、精度が悪くなってしまうこともある。

次数を増やす
次数を増やしていくと数式は、１次式：ax+b 2次式：ax^2+bx+c 3次式：ax^3+bx^2+cx+d・・・　のように複雑になっていく。
次数を増やすことにより、データへのフィッティングの精度は上がるが、外れ値などにも過剰にフィットしてしまうなど過学習しやすく、新たなデータを当てはめたときの予測の精度は落ちる可能性がある。

重回帰分析
説明変数を増やすことで、精度の高いモデルを作ることができる。増やし過ぎると、オーバーフィッティングしやすく、また多重共線性が生じることで予測の精度が落ちることに繋がってしまう。
